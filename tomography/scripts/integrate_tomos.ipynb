```python
import pandas as pd
import os
```

```python
# list the csv files in ../data/*/tomography_data.csv directory
data_dir = '../data'
csv_files = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file == 'tomography_data.csv':
            csv_files.append(os.path.join(root, file))

# remove the directory "old_tomos" from the list
csv_files = [f for f in csv_files if 'old_tomos' not in f]
csv_files
```

```python
# for each file, read the csv, add together the coumns "TT" and "RT", rename that to "T", add together the columns "TR" and "RR", rename that to "R", and save the result to a new csv file in the same directory
for file in csv_files:
    df = pd.read_csv(file)
    df['T'] = df['TT'] + df['RT']
    df['R'] = df['TR'] + df['RR']
    df = df
    new_file_name = file.replace('tomography_data.csv', 'integrated_tomography_data.csv')
    df.to_csv(new_file_name, index=False)
    print(f"Processed {file} and saved to {new_file_name}")
```
